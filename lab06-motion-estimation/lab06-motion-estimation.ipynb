### Introduction

This notebook introduces semantic segmentation using the Oxford Pets dataset. Semantic segmentation assigns a class label to each pixel in an image, for example distinguishing the pet from the background.

#### What is Optical Flow?
Optical flow is a task of per-pixel motion estimation between two consecutive frames in one video. Basically, the Optical Flow task implies the calculation of the shift vector for pixel as an object displacement difference between two neighboring images. The main idea of Optical Flow is to estimate the object’s displacement vector caused by it’s motion or camera movements.

#### Sparse vs Dense Optical Flow
**1. Sparse Optical Flow**  
- Tracks only key points (corners/features)
- **Example:** Lucas–Kanade
- **Advantages:** Fast, lightweight
- **Disadvantages:** Limited detail
- **Applications:** Object Tracking

**2. Dense Optical Flow**  
- Computes motion for every pixel in the image.
- **Examples:** Farneback, RAFT, RLOF
- **Advantages:** Detailed motion
- **Disadvantages:** Slower, heavy
- **Applications:** Full scene analysis


### What is a Flow Map?
A flow map is a visual representation of optical flow, it shows how every pixel (or selected points) moves between two frames in a video.

It’s essentially a map of motion.

#### How It Works
- Each pixel is assigned a motion vector (dx, dy)
- Vectors show direction and speed of movement
- Can be visualized in different ways:
    1. **Arrows**: Each arrow represents the **direction and magnitude** of motion at a pixel or selected point.  
        Commonly used in **sparse optical flow**.

    2. **Color-coded Map**: Uses **hue to indicate direction** and **brightness to indicate the magnitude (speed) of the motion**.  
        Commonly used in **dense optical flow**

#### Running code locally
1. Navigate to the **Optical-Flow-Estimation-using-Deep-Learning-RAFT** folder

    `cd "Optical-Flow-Estimation-using-Deep-Learning-RAFT"`

2. To run the demo, firstly you need to clone the RAFT repo being in our directory:

    `git clone https://github.com/princeton-vl/RAFT.git`


3. Create a new **virtual environment**:

    ```
    conda create --name raft python=3.12
    conda activate raft
    ```

4. Install the required libraries:

    `pip install -r requirements.txt`

5. Now you can run the demo with RAFT:

    `python inference.py --model=./models/raft-sintel.pth --video ./videos/crowd.mp4`

    or use the Raft Small Model

    `python inference.py --model=./models/raft-small.pth --video ./videos/crowd.mp4 --small`
